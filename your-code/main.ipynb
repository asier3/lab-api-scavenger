# Challenge 2

```python
response = requests.get('https://api.github.com/repos/ta-data-lis/lab-list-comprehensions/events')

response.status_code
```

```python
events = pd.json_normalize(response.json())

events.head(3)
```

```python
# We will filter the 'Pull Request events' in order to find the pulls
condition = events['type'] == 'PullRequestEvent'

pulls = events[condition]
pulls.head(3)
```

```python
# Will extract the date and pass it to 'yyyymmdd' format and create a rule to filter the pulls in the last 2 weeks.

def extract_date (row):
    return row['created_at'][:10].replace('-', '')

pulls['extr_date'] = pulls.apply(extract_date, axis=1)
pulls.head(3)
```

```python
pulls = pulls.astype({"extr_date": int}) 
```

```python
# In the following dataframe we can see the pulls that happened the last 2 weeks
condition = pulls['extr_date'] >= 20220418
pulls_filtered = pulls[condition]
pulls_filtered
```

```python
# We can show the data properly now, and this is our result

pulls_clean = pulls_filtered[['id','type','actor.display_login','created_at']].reset_index()
pulls_clean
```

```python
# We can also count the values 

pulls_clean.groupby('type').agg({'index':'count'})
```
